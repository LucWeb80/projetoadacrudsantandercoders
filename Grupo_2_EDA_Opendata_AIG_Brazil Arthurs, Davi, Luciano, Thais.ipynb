## Quais são os tipos de Ocorrências Mais Recorrentes em `Acidentes Fatais` e `Não Fatais`


```python


df['tipo_acidente'] = df['quantidade_fatalidades'].apply(lambda x: 'Fatal' if x > 0 else 'Não Fatal')

# Filtrar dados para acidentes fatais e não fatais
acidentes_fatais = df[df['tipo_acidente'] == 'Fatal']
acidentes_nao_fatais = df[df['tipo_acidente'] == 'Não Fatal']

# Contar os tipos de ocorrências mais recorrentes para ambos os grupos
tipos_fatais = acidentes_fatais['ocorrencia_tipo'].value_counts().head(5)
tipos_nao_fatais = acidentes_nao_fatais['ocorrencia_tipo'].value_counts().head(5)


print("Tipos de Ocorrências Mais Recorrentes em Acidentes Fatais:")
print(tipos_fatais)
print("\nTipos de Ocorrências Mais Recorrentes em Acidentes Não Fatais:")
print(tipos_nao_fatais)
```

- Foi possível observar que a perda de controle em voo é a maior causa de acidentes fatais, aparecendo `48%` dos casos

- E em ocorrências não fatais, não existe algum padrão evidente ou valor muito diferente dos demais

## Tipos de Ocorrências Mais Recorrentes em `Acidentes Fatais` e `Não Fatais`

```python


df['tipo_acidente'] = df['quantidade_fatalidades'].apply(lambda x: 'Fatal' if x > 0 else 'Não Fatal')


tipos_fatais = df[df['tipo_acidente'] == 'Fatal']['ocorrencia_tipo'].value_counts().head(5)
tipos_nao_fatais = df[df['tipo_acidente'] == 'Não Fatal']['ocorrencia_tipo'].value_counts().head(5)


tipos_combinados = pd.DataFrame({
    'Fatal': tipos_fatais,
    'Não Fatal': tipos_nao_fatais
}).fillna(0)


plt.figure(figsize=(18, 14))
ax = tipos_combinados.plot(kind='barh', width=0.6, color=['red', 'green'], edgecolor='black')


plt.subplots_adjust(left=0.25, right=0.75, top=0.85, bottom=0.2)


plt.suptitle('Tipos de Ocorrências Mais Recorrentes em Acidentes Fatais e Não Fatais', fontsize=22, x=0.5, y=0.95, ha='center')
plt.xlabel('Número de Ocorrências', fontsize=18)
plt.ylabel('Tipo de Ocorrência', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


labels = [label.get_text() for label in ax.get_yticklabels()]
ax.set_yticklabels(labels)

# Mover a legenda para fora do gráfico
plt.legend(title='Tipo de Acidente', fontsize=14, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

- Após a visualização do gráfico, é possível observar que algumas ocorrências em nenhum dos casos foram classificadas como não fatais. São elas:

 `Perda de controle em voo`

  `Colisão com obstáculo durante a decolagem e pouso`


- Assim como há ocorrências em que nenhuma delas foi classificada como fatal. São elas:

  `Estouro de Pneu`

  `Falha ou Mau funcionamento de sistema/componente`

  `Perda de controle no solo`

```python

# Dicionário para mapear os estados brasileiros às suas regiões
regioes = {
    'AC': 'Norte', 'AL': 'Nordeste', 'AP': 'Norte', 'AM': 'Norte', 'BA': 'Nordeste',
    'CE': 'Nordeste', 'DF': 'Centro-Oeste', 'ES': 'Sudeste', 'GO': 'Centro-Oeste',
    'MA': 'Nordeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste', 'MG': 'Sudeste',
    'PA': 'Norte', 'PB': 'Nordeste', 'PR': 'Sul', 'PE': 'Nordeste', 'PI': 'Nordeste',
    'RJ': 'Sudeste', 'RN': 'Nordeste', 'RS': 'Sul', 'RO': 'Norte', 'RR': 'Norte',
    'SC': 'Sul', 'SP': 'Sudeste', 'SE': 'Nordeste', 'TO': 'Norte'
}

data = pd.read_csv('accidents.csv')
data['regiao'] = data['ocorrencia_uf'].map(regioes)


data_brasil = data.dropna(subset=['regiao'])

ocorrencias_por_regiao = data_brasil.groupby('regiao').size()

ocorrencias_graves_por_regiao = data_brasil[data_brasil['aeronave_nivel_dano'].isin(['SUBSTANCIAL', 'DESTRUÍDA'])].groupby('regiao').size()

ocorrencias_fatais_por_regiao = data_brasil.groupby('regiao')['quantidade_fatalidades'].sum()
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico 1: Percentual do número de ocorrências por região
axs[0].pie(ocorrencias_por_regiao, labels=ocorrencias_por_regiao.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
axs[0].set_title('Ocorrências por Região', fontsize=14)

# Gráfico 2: Percentual de ocorrências graves por região
axs[1].pie(ocorrencias_graves_por_regiao, labels=ocorrencias_graves_por_regiao.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
axs[1].set_title('Ocorrências Graves por Região', fontsize=14)

# Gráfico 3: Percentual de acidentes fatais por região
axs[2].pie(ocorrencias_fatais_por_regiao, labels=ocorrencias_fatais_por_regiao.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
axs[2].set_title('Acidentes Fatais por Região', fontsize=14)

# Ajustar o layout para melhor visualização
plt.tight_layout()
plt.show()
```

## <center> Análise dos Gráficos

1. **Ocorrências por Região**:
   - A maioria das ocorrências de acidentes aéreos está concentrada na região Sudeste.
   - Outras regiões como Centro-Oeste, Nordeste, Norte e Sul têm uma menor proporção de ocorrências.

2. **Ocorrências Graves por Região**:
   - A região Sudeste ainda apresenta uma alta porcentagem de ocorrências graves, mas o Centro-Oeste também tem uma proporção significativa.
   - Isso pode indicar que, embora o Sudeste tenha mais acidentes em geral, a gravidade dos acidentes no Centro-Oeste é relativamente alta.

3. **Acidentes Fatais por Região**:
   - A região Norte apresenta uma alta porcentagem de acidentes fatais, apesar de não ter tantas ocorrências gerais.
   - O Sudeste, embora tenha muitas ocorrências e ocorrências graves, tem uma proporção menor de acidentes fatais em comparação com o Norte.

### Conclusão

A localização geográfica influencia tanto o tipo quanto a gravidade dos acidentes aéreos. Algumas observações importantes incluem:

- **Tráfego Aéreo**: Regiões com maior tráfego aéreo, como o Sudeste, tendem a ter mais ocorrências de acidentes.
- **Condições Locais**: Fatores como condições meteorológicas, geografia e infraestrutura podem influenciar a gravidade dos acidentes. Por exemplo, o Norte pode ter condições mais desafiadoras que resultam em uma maior proporção de acidentes fatais.
- **Infraestrutura e Resposta**: A capacidade de resposta a emergências e a qualidade da infraestrutura aeroportuária também podem variar entre as regiões, afetando a gravidade dos acidentes.

## Qual é a distribuição das classificações das ocorrências (acidente, incidente, incidente grave)?

```python

classificacao_ocorrência = df['ocorrencia_classificacao'].value_counts()


classificacao_ocorrência
```

```python

total = classificacao_ocorrência.sum()
proporcoes = classificacao_ocorrência / total * 100

plt.figure(figsize=(10, 6))
ax = classificacao_ocorrência.plot(kind='bar', color='skyblue')


for i, v in enumerate(classificacao_ocorrência):
    height = v
    x = i
    y = height
    proporcao = proporcoes[i]
    ax.text(x, y, f'{proporcao:.1f}%', ha='center', va='bottom')


plt.title('Distribuição das Classificações de Ocorrências')
plt.xlabel('Classificação da Ocorrência')
plt.ylabel('Número de Ocorrências')


plt.show()
```

 ### De acordo com a Agencia Nacional de Aviacao Civil (`ANAC`) podemos classificar da seguinte forma


###- `Acidente Aeronáutico`: Ocorre quando há danos significativos a uma aeronave, resultando em lesões graves ou morte para as pessoas a bordo.




###- `Incidente Grave`: Representa um risco significativo para a segurança, mas não atinge o nível de um acidente.



###- `Incidente Aeronáutico`: Afeta a segurança da operação, mas não chega a ser grave.

## Quais são os tipos mais comuns de ocorrências relatadas?

```python

# Filtrar os dados para excluir a categoria "OUTROS"
dados_filtrados = df[df['ocorrencia_tipo'] != 'OUTROS']


ocorrencia_count = dados_filtrados['ocorrencia_tipo'].value_counts().head(10)

# Ordenar os tipos de ocorrência em ordem decrescente
ocorrencia_count = ocorrencia_count.sort_values(ascending=True)


plt.figure(figsize=(10, 6))
ax = ocorrencia_count.plot(kind='barh',color='#5B88A1')


plt.title('Os 10 tipos de ocorrências mais comuns')
plt.xlabel('Número de Ocorrências')
plt.ylabel('Tipo de Ocorrência')
plt.xticks(rotation=0)

# Adicionar porcentagem nas barras
total = ocorrencia_count.sum()
for p in ax.patches:
    width = p.get_width()
    percentage = (width / total) * 100
    x = width + 10  # Deslocamento para a direita
    y = p.get_y() + p.get_height() / 2
    ax.text(x, y, f'{percentage:.1f}%', va='center')


plt.tight_layout()
plt.show()
```

###Conforme observado na análise realizada por Arthur Gomes, juntamente com este gráfico, podemos perceber que os tipos de ocorrência que ocorrem com maior frequência são aqueles com poucas ou nenhuma fatalidade

### Qual a correlação entre a idade da aeronave e a quantidade de acidentes?




```python
df = df[df['aeronave_ano_fabricacao'] > 0].copy()
df['idade_aeronave'] = 2024 - df['aeronave_ano_fabricacao']
df_acidente = df[df['ocorrencia_classificacao'] == 'ACIDENTE'].copy()
acidente_groupby = df_acidente.groupby(['idade_aeronave']).size().sort_values(ascending=False)
```

```python
acidente_groupby.reset_index(level='idade_aeronave').rename(columns={0: 'n_ocorrencias'}).corr(method='spearman')
```

- Existe correlação negativa entre a quantidade de ocorrências e a idade das aeronaves

- Mas não necessariamente aeronaves mais novas tendem a dar mais problemas, mas sim a quantidade de aeronaves e voos que aumentou durante o período de 10 anos


### Distribuição de Acidentes por Tipo e Quantidade de Motores

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_1 = df[df['aeronave_tipo_motor'] != '***'].copy()
grouped = df_1.groupby(['aeronave_tipo_motor', 'aeronave_quantidade_motores']).size().unstack().fillna(0)

ax = grouped.plot(kind='bar', stacked=True, figsize=(12, 8))
ax.set_title('Distribuição de Acidentes por Tipo e Quantidade de Motores', fontsize=16)
ax.set_xlabel('Tipo de Motor', fontsize=14)
ax.set_ylabel('Número de Acidentes', fontsize=14)
ax.legend(title='Quantidade de Motores', title_fontsize='13', fontsize='11')

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    if height > 10:
        ax.text(x + width/2, y + height/2, int(height), ha='center', va='center', fontsize=10, color='black')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
